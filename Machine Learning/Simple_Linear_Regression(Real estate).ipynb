```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('Real estate.csv')

# Select a feature for simple linear regression (e.g., house age)
X = df[['X2 house age']]
y = df['Y house price of unit area']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Create and train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate R² scores as percentages
train_r2_percentage = r2_score(y_train, y_train_pred) * 100
test_r2_percentage = r2_score(y_test, y_test_pred) * 100

# Plot the training data and the model
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_train_pred, color='red', label='Linear regression')
plt.xlabel('House Age')
plt.ylabel('House Price per Unit Area')
plt.title('House Age vs. House Price')
plt.legend()
plt.show()

# Evaluate the model
train_r2 = r2_score(y_train, y_train_pred)
test_r2 = r2_score(y_test, y_test_pred)
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Training Mean Squared Error (MSE):', train_mse)
print('Testing Mean Squared Error (MSE):', test_mse)
print('Training Coefficient of Determination (R^2 score):', train_r2)
print('Testing Coefficient of Determination (R^2 score):', test_r2)
print(f'Training R² Score: {train_r2_percentage:.2f}%')
print(f'Testing R² Score: {test_r2_percentage:.2f}%')
```
